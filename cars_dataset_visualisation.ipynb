```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cars.csv')
df.head()
```

```python
df.info()
```

```python
#convert cubicinches and weightlbs to numeric datatype(float)

df['cubicinches']= pd.to_numeric(df['cubicinches'], errors='coerce')
df['weightlbs']= pd.to_numeric(df['weightlbs'], errors='coerce')

#error='corece' : will automatically take non-convertible or missing values as NaN

#pd.to_datetime : fxn to convert any column into date time format
```
